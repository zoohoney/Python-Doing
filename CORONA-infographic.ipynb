```python
#리스트에 20대확진자 수 80대 확진자 수 만듬
with open('/home/aiffel-dj31/CORONA/TimeAge.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    tw = []
    eg = []
    may = []
    for row in data :
        if row[0].split('-')[1] == '05':
            if row[2] == '20s' :
                tw.append(row[3])
                may.append(row[0].split('-')[-1])
            elif row[2] == '80s' :
                eg.append(row[3])
            
print('20대',tw)
print('80대',eg)
print('날짜',may)
```

```python
# 시각화 해보기 
plt.figure(figsize=(15,10))
plt.title('5월 20대와80대 확진자수')
plt.plot(may,tw,'blue', label='male')
plt.plot(eg,'red', label='female')
plt.legend()
plt.show()
```

```python
with open('/home/aiffel-dj31/CORONA/TimeAge.csv',encoding ='cp949') as f:
    data = csv.reader(f)
    next(data)
    tw = []
    eg = []
    may = []
    for row in data :
        if row[0].split('-')[1] == '05':
            if row[2] == '20s' :
                tw.append(int(row[3]))
                may.append(row[0].split('-')[-1])
            elif row[2] == '80s' :
                eg.append(-int(row[3]))
            
print('20대',tw)
print('80대',eg)
print('날짜',may)
```

```python
# 항아리 만들어 봣는데 흐음....
plt.rcParams['axes.unicode_minus'] = False
plt.figure(figsize=(20,10))
plt.title('5월 20대와 80대 ')
plt.barh(may,tw,color="r",label="20대")
plt.barh(may,eg,color="b",label="80대")
plt.legend()
plt.show()
```

```python
import matplotlib.pyplot as plt

for i in range(1,5):
    plt.figure(figsize= (20,15))
    plt.style.use('ggplot')      # 격자 무늬 스타일 지정

    # 20대 확진자 시각화
    plt.subplot(4,2,2*i-1)
    title='20년'+str(i+2)+"월 20대 일별 확진자"
    plt.title(title,fontsize=15)
    plt.xticks( fontsize=15)
    plt.yticks( fontsize=15)
    for j in range(len(month_date_date[i-1])):
        plt.bar(month_date_date[i-1][j],month_date_confirmed[i-1][j],color='b')
    #plt.rc('legend', fontsize=20) # 라벨 사이즈
    #plt.legend()

    # 80대 확진자 시각화
    plt.subplot(4,2,i*2)
    title='20년'+str(i+2)+"월 80대 일별 확진자"
    plt.title(title,fontsize=15)
    plt.xticks( fontsize=15)
    plt.yticks( fontsize=15)
    for j in range(len(month_date_date[i-1])):
        plt.bar(month_date_date[i-1][j],month_date_confirmed80[i-1][j],color='pink')
    
plt.show()
```
